```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('earth_meteorites.csv')

# Convert the 'year' column to datetime format
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# Get all the Earth meteorites that fell before the year 2000
before_2000 = df[df['year'] < pd.to_datetime('2000-01-01')]
print("Earth meteorites that fell before the year 2000:")
print(before_2000)

# Get all the Earth meteorites coordinates that fell before the year 1970
before_1970 = df[df['year'] < pd.to_datetime('1970-01-01')]
print("Earth meteorites coordinates that fell before the year 1970:")
print(before_1970[['reclat', 'reclong']])

# Plotting scatter plot of meteorite coordinates
plt.scatter(df['reclong'], df['reclat'], marker='o', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorite Fall Coordinates')
plt.show()
```
