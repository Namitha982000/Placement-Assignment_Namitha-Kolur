```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

# Define and compile the first CNN architecture
model_1 = tf.keras.Sequential([
    layers.Conv2D(32, (3, 3), activation="relu", input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])
model_1.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Train the first CNN architecture
model_1.fit(x_train, y_train, batch_size=128, epochs=10, validation_data=(x_test, y_test))

# Define and compile the second CNN architecture
model_2 = tf.keras.Sequential([
    layers.Conv2D(16, (3, 3), activation="relu", input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])
model_2.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Train the second CNN architecture
model_2.fit(x_train, y_train, batch_size=128, epochs=10, validation_data=(x_test, y_test))

# Define and compile the third CNN architecture
model_3 = tf.keras.Sequential([
    layers.Conv2D(32, (5, 5), activation="relu", input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (5, 5), activation="relu"),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(256, activation="relu"),
    layers.Dense(10, activation="softmax")
])
model_3.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Train the third CNN architecture
model_3.fit(x_train, y_train, batch_size=128, epochs=10, validation_data=(x_test, y_test))

# Evaluate the models on the test set
score_1 = model_1.evaluate(x_test, y_test, verbose=0)
score_2 = model_2.evaluate(x_test, y_test, verbose=0)
score_3 = model_3.evaluate(x_test, y_test, verbose=0)

# Create a comparison table
table = [["Architecture", "Accuracy", "Parameters"],
         ["Model 1", f"{score_1[1]*100:.2f}%", model_1.count_params()],
         ["Model 2", f"{score_2[1]*100:.2f}%", model_2.count_params()],
         ["Model 3", f"{score_3[1]*100:.2f}%", model_3.count_params()]]

# Print the comparison table
print("{:<12} {:<10} {:<12}".format(*table[0]))
for row in table[1:]:
    print("{:<12} {:<10} {:<12}".format(*row))
```
